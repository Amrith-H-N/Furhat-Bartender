# data processing using pyfeat

```python
# Import necessary libraries pillow, plotly needs to be installed
import imageio.v3 as iio
import cv2
from matplotlib import pyplot as plt
from IPython.display import Image, Video
from feat import Detector
import opencv_jupyter_ui as jcv2
import torch
import pandas as pd
import os
from PIL import Image, ImageDraw, ImageFont
```

```python
# Set up paths and read the CSV files with proper parameters
#images_path = "../dataset/DiffusionEmotion_S/cropped/all/"
images_path = ["../dataset/DiffusionEmotion_S/cropped/angry/","../dataset/DiffusionEmotion_S/cropped/disgust/","../dataset/DiffusionEmotion_S/cropped/fear/","../dataset/DiffusionEmotion_S/cropped/happy/","../dataset/DiffusionEmotion_S/cropped/neutral/","../dataset/DiffusionEmotion_S/cropped/sad/","../dataset/DiffusionEmotion_S/cropped/surprise/"]
```

```python
detector = Detector(device="cuda")
```

```python
dest_directory = "../processed/"

emo_text = ["anger", "disgust", "fear", "happiness", "sadness", "surprise", "neutral"]
first = True
for img in images_path :
    for file_name in os.listdir(img):
        if file_name.endswith(".png"):
        # print(file_name)
            face = 0

            prediction = detector.detect_image(img + str(file_name))
            emotion_name = img.split("/")[-2]
            prediction['emotions'] = emotion_name
            print("emotion name: '{emotion_name}':")

        # prediction.head()
            rows = len(prediction)

            if first:
                au_activations = pd.DataFrame(prediction)
                first = False
            else:
                au_activations = pd.concat([au_activations, prediction])
```

# save the data in processed/aus.csv

```python
au_activations.reset_index()
au_activations.to_csv(dest_directory + "aus.csv", sep=",", encoding="utf-8")
df = pd.read_csv(dest_directory + "aus.csv")
df.rename(columns={"Unnamed: 0": "face"}, inplace=True)
df.to_csv(dest_directory + "aus.csv", sep=",", encoding="utf-8")
```
